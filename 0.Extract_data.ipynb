```python
import pandas as pd
from zipfile import ZipFile

directory = '/content/drive/MyDrive/Thesis'


data_dir = directory + "/Data/Baywheels"

import glob

zip_files = glob.glob(f"{data_dir}/*.zip")
```

```python
import os
import requests
import zipfile
from io import BytesIO

def download_and_extract_zip(url, extract_to='.'):
    # Download the file from the given URL
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Read the ZIP file in memory
        with zipfile.ZipFile(BytesIO(response.content)) as zip_ref:
            # Extract all contents to the specified directory
            zip_ref.extractall(extract_to)
        print(f"Files extracted to: {os.path.abspath(extract_to)}")
    else:
        print(f"Failed to download the file. Status code: {response.status_code}")

# Example usage
for i in ['202201', '202202', '202203', '202204', '202205', '202206', '202207','202208','202209','202210','202211']:
  url = f'https://s3.amazonaws.com/baywheels-data/{i}-baywheels-tripdata.csv.zip'  # Replace with the actual URL
  extract_to_directory = data_dir   # Replace with the actual directory path
  download_and_extract_zip(url, extract_to_directory)
```

```python
for zip_file in zip_files:
  with ZipFile(zip_file, 'r') as zObject:

      # Extracting all the members of the zip
      # into a specific location.
      zObject.extractall(path=data_dir)
```
